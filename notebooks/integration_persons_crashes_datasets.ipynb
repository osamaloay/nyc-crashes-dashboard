```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# Paths (adjust to your exact filenames)
mine = '/content/drive/MyDrive/df_crashes_geocoded_final.csv'
college = '/content/drive/MyDrive/cleaned_persons_data.csv'

df_mine = pd.read_csv(mine)
df_college = pd.read_csv(college)
```

```python
print("Mine shape:", df_mine.shape)
print("College shape:", df_college.shape)

print("\nCommon columns:", set(df_mine.columns) & set(df_college.columns))
print("\nColumns only in mine:", set(df_mine.columns) - set(df_college.columns))
print("\nColumns only in college:", set(df_college.columns) - set(df_mine.columns))
```

```python
merged = pd.merge(df_college, df_mine, on='COLLISION_ID', how='left')
```

```python
# Show all column names
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)  # optional: just show top 20 rows

merged.head()

```

```python
# Crash-level
crash_cols = [
    'COLLISION_ID','CRASH_DATETIME_x','BOROUGH','ZIP CODE','LATITUDE','LONGITUDE',
    'ON STREET NAME','OFF STREET NAME','CROSS STREET NAME','ALL_VEHICLE_TYPES',
    'NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED',
    'NUMBER OF PEDESTRIANS INJURED','NUMBER OF PEDESTRIANS KILLED',
    'NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED',
    'NUMBER OF MOTORIST INJURED','NUMBER OF MOTORIST KILLED',
    'ALL_CONTRIBUTING_FACTORS_STR','FULL ADDRESS'
]

# Person-level
person_cols = [c for c in merged.columns if c not in crash_cols]
```

```python
# If both _x and _y exist, prefer mine (_y)
merged['CRASH_DATETIME'] = merged['CRASH_DATETIME_y'].combine_first(merged['CRASH_DATETIME_x'])
merged.drop(columns=['CRASH_DATETIME_x','CRASH_DATETIME_y'], inplace=True)
```

```python
position_map = {
    'Driver': 'DRIVER',

    'Front passenger, if two or more persons, including the driver, are in the front seat':
        'FRONT PASSENGER',

    'Middle front seat, or passenger lying across a seat':
        'MIDDLE FRONT',

    'Left rear passenger, or rear passenger on a bicycle, motorcycle, snowmobile':
        'REAR LEFT',

    'Middle rear seat, or passenger lying across a seat':
        'REAR MIDDLE',

    'Right rear passenger or motorcycle sidecar passenger':
        'REAR RIGHT',

    'Riding/Hanging on Outside':
        'OUTSIDE VEHICLE',

    'If one person is seated on another person&apos;s lap':
        'LAP SEATED',

    'Any person in the rear of a station wagon, pick-up truck, all passengers on a bus, etc':
        'OTHER REAR',

    'Does Not Apply':
        'NOT APPLICABLE',

    'Unknown':
        'UNKNOWN',
}
```

```python
merged['POSITION_IN_VEHICLE_CLEAN'] = (
    merged['POSITION_IN_VEHICLE']
    .map(position_map)
    .fillna('OTHER')
)
```

```python
merged.drop(columns=['POSITION_IN_VEHICLE'], inplace=True)
```

```python
crash_cols = [
    'BOROUGH','ZIP CODE','LATITUDE','LONGITUDE',
    'ON STREET NAME','OFF STREET NAME','CROSS STREET NAME'
]

# Count unique values for each collision in each crash-level column
check = (
    merged.groupby("COLLISION_ID")[crash_cols]
    .nunique(dropna=False)
)

check.head()
```

```python
inconsistent = check[(check > 1).any(axis=1)]
print("Number of inconsistent collisions:", inconsistent.shape[0])
inconsistent.head()
```

```python
# Drop collisions where all critical crash-level columns are NaN
critical_cols = ['BOROUGH','ZIP CODE','LATITUDE','LONGITUDE',
                 'ON STREET NAME','OFF STREET NAME','CROSS STREET NAME']

# Keep only collisions with at least one non-null value in critical columns
merged_clean = merged.dropna(subset=critical_cols, how='all').reset_index(drop=True)
print("Shape after dropping fully empty collisions:", merged_clean.shape)
```

```python
critical_cols = ['BOROUGH','ZIP CODE','LATITUDE','LONGITUDE',
                 'ON STREET NAME','OFF STREET NAME','CROSS STREET NAME']

# Fill missing values **only if some value exists in the collision group**
for col in critical_cols:
    merged[col] = merged.groupby('COLLISION_ID')[col].transform(
        lambda x: x.fillna(method='ffill').fillna(method='bfill')
    )


print("Shape after filling and keeping valid collisions:", merged_clean.shape)
```

```python
FINAL_PATH = "/content/drive/MyDrive/merged_cleaned_dataset.csv"
merged_clean.to_csv(FINAL_PATH, index=False)
print("Saved final dataset to:", FINAL_PATH)
```
