```python
import re

vehicle_patterns = {
    'Sedan': [r'\b(sedan|4 dr|2 dr)\b'],
    'SUV/Station Wagon': [r'\b(suv|station wagon|sport utility|suv/sta)\b'],
    'Pickup Truck': [r'\b(pick[- ]?up|pickup|pick up truck)\b'],
    'Truck/Commercial': [r'\b(truck|tractor truck|flatbed|semi|delivery|box truck|commercial|bulk|garbage|tow[- ]?truck|dump|concrete|milk|beverage)\b'],
    'Van': [r'\b(van|van camper|refrigerated|utility van|work van|uhaul)\b'],
    'Motorcycle': [r'\b(motorcycle|motorbike|moped|moped/scooter|scooter|e-bike|e-bike|ebike|motorscooter|dirt bike)\b'],
    'Bus': [r'\b(bus|school bus|mta|nyc transit|omnibus|coach)\b'],
    'Taxi': [r'\b(taxi|livery)\b'],
    'Ambulance': [r'\b(ambulance|fdny|nyc ambula)\b'],
    'Pedicab': [r'\b(pedicab)\b'],
    'Bicycle': [r'\b(bike|bicycle|unicycle)\b'],
    'Other': [r'.*']  # fallback
}
```

```python
def clean_vehicle_type(raw):
    if not isinstance(raw, str):
        return 'Other'
    raw_lower = raw.lower()
    for category, patterns in vehicle_patterns.items():
        for pat in patterns:
            if re.search(pat, raw_lower):
                return category
    return 'Other'
```

```python
import pandas as pd

street_cols = ['ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME']


def clean_street_name(name):
    if pd.isna(name):
        return None
    name = str(name).upper().strip()
    name = name.replace('.', '')
    name = name.replace('-', ' ')
    name = ' '.join(name.split())          # Collapse multiple spaces into one

    # Expand common abbreviations
    replacements = {
        ' ST ': ' STREET ',
        ' AV ': ' AVENUE ',
        ' AVE ': ' AVENUE ',
        ' BLVD ': ' BOULEVARD ',
        ' RD ': ' ROAD ',
        ' PKWY ': ' PARKWAY ',
        ' PL ': ' PLACE ',
        ' TER ': ' TERRACE ',
        ' HWY ': ' HIGHWAY ',
        ' LN ': ' LANE ',
        ' SQ ': ' SQUARE ',
        ' DR ': ' DRIVE ',
        ' CTR ': ' CENTER ',
        ' LOOP ': ' LOOP ',
        ' CIR ': ' CIRCLE '
    }

    for k, v in replacements.items():
        name = name.replace(k, v)

    return name


for col in street_cols:
    df_crashes[col] = df_crashes[col].apply(clean_street_name)


df_crashes[street_cols].head(10)
```

```python
# Function to combine streets into one address string
def build_address(row):
    parts = []
    if row['ON STREET NAME']:
        parts.append(row['ON STREET NAME'])
    if row['CROSS STREET NAME']:
        parts.append('AND ' + row['CROSS STREET NAME'])
    if row['OFF STREET NAME']:
        parts.append(row['OFF STREET NAME'])
    return ', '.join(parts) if parts else None

df_crashes['FULL ADDRESS'] = df_crashes.apply(build_address, axis=1)

# Optional: append city and state to help geocoding
df_crashes['FULL ADDRESS'] = df_crashes['FULL ADDRESS'].apply(
    lambda x: f"{x}, New York, NY" if x else None
)

df_crashes[['ON STREET NAME','CROSS STREET NAME','OFF STREET NAME','FULL ADDRESS']].head(10)
```

```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time
import os
import geopandas as gpd
from shapely.geometry import Point
from google.colab import drive

# -----------------------------
# 0. Mount Google Drive (Colab)
# -----------------------------
drive.mount('/content/drive')

SAVE_DIR = '/content/drive/MyDrive/geocoding_batches'
FINAL_SAVE = '/content/drive/MyDrive/df_crashes_geocoded.csv'
os.makedirs(SAVE_DIR, exist_ok=True)

# -----------------------------
# 1. Initialize geocoder
# -----------------------------
geolocator = Nominatim(user_agent="nyc_crash_geocoder")
geocode = RateLimiter(
    geolocator.geocode,
    min_delay_seconds=1,
    max_retries=2,
    error_wait_seconds=2.0,
    swallow_exceptions=True
)

# -----------------------------
# 2. Filter rows that need geocoding
# -----------------------------
df_to_geocode = df_crashes[
    df_crashes['LATITUDE'].isna() | df_crashes['LONGITUDE'].isna()
].copy()

df_to_geocode = df_to_geocode[df_to_geocode['FULL ADDRESS'].notna()]
print(f"Total rows to geocode: {len(df_to_geocode):,}")

# -----------------------------
# 3. Resume support
# -----------------------------
processed_batches = [f for f in os.listdir(SAVE_DIR) if f.startswith("geocoded_batch_")]
start_batch = len(processed_batches)
print(f"Resuming from batch {start_batch} (found {len(processed_batches)} saved files).")

# -----------------------------
# 4. Batch geocoding loop
# -----------------------------
BATCH_SIZE = 500
num_batches = (len(df_to_geocode) + BATCH_SIZE - 1) // BATCH_SIZE

for batch_num in range(start_batch, num_batches):
    batch_file = os.path.join(SAVE_DIR, f'geocoded_batch_{batch_num}.csv')

    start_idx = batch_num * BATCH_SIZE
    end_idx = min((batch_num + 1) * BATCH_SIZE, len(df_to_geocode))
    df_batch = df_to_geocode.iloc[start_idx:end_idx].copy()

    if df_batch.empty:
        continue

    print(f"\nðŸ“¦ Processing batch {batch_num+1}/{num_batches} ({len(df_batch)} rows)...")

    # Step-by-step geocoding to handle errors better
    latitudes, longitudes = [], []
    for i, address in enumerate(df_batch['FULL ADDRESS']):
        try:
            location = geocode(address)
            lat = location.latitude if location else None
            lon = location.longitude if location else None
        except Exception:
            lat, lon = None, None

        latitudes.append(lat)
        longitudes.append(lon)

        # Save partial every 50 addresses
        if (i + 1) % 50 == 0:
            temp_df = pd.DataFrame({
                'FULL ADDRESS': df_batch['FULL ADDRESS'].iloc[:i+1],
                'LATITUDE_GEOCODE': latitudes,
                'LONGITUDE_GEOCODE': longitudes
            })
            temp_path = os.path.join(SAVE_DIR, f'batch_{batch_num}_partial.csv')
            temp_df.to_csv(temp_path, index=False)
            print(f"ðŸ’¾ Saved partial {i+1}/{len(df_batch)} for batch {batch_num+1}")

    # Save batch results
    df_batch['LATITUDE_GEOCODE'] = latitudes
    df_batch['LONGITUDE_GEOCODE'] = longitudes
    df_batch[['FULL ADDRESS','LATITUDE_GEOCODE','LONGITUDE_GEOCODE']].to_csv(batch_file, index=False)
    print(f"âœ… Saved full batch {batch_num+1} â†’ {batch_file}")

    # Merge into main df_crashes
    df_crashes.loc[df_batch.index, 'LATITUDE'] = df_crashes.loc[df_batch.index, 'LATITUDE'].combine_first(df_batch['LATITUDE_GEOCODE'])
    df_crashes.loc[df_batch.index, 'LONGITUDE'] = df_crashes.loc[df_batch.index, 'LONGITUDE'].combine_first(df_batch['LONGITUDE_GEOCODE'])

    # Save progress to disk
    df_crashes.to_csv(FINAL_SAVE, index=False)
    print(f"ðŸ“‚ Progress saved to {FINAL_SAVE}")

    time.sleep(1)

print("\nðŸŽ‰ All batches processed successfully.")

# -----------------------------
# 5. Spatial join: fill borough + zip
# -----------------------------
print("\nðŸ§­ Performing spatial join to infer BOROUGH and ZIP CODE...")

# Convert to GeoDataFrame
crashes_gdf = gpd.GeoDataFrame(
    df_crashes.dropna(subset=['LATITUDE', 'LONGITUDE']),
    geometry=gpd.points_from_xy(df_crashes['LONGITUDE'], df_crashes['LATITUDE']),
    crs="EPSG:4326"
)

# Fill missing BOROUGH
crashes_with_borough = gpd.sjoin(
    crashes_gdf, boroughs[['BoroName', 'geometry']], how='left', predicate='within'
)
df_crashes['BOROUGH'] = df_crashes['BOROUGH'].combine_first(
    crashes_with_borough.set_index(crashes_with_borough.index)['BoroName']
)

# Fill missing ZIP CODE
crashes_with_zip = gpd.sjoin(
    crashes_gdf, zipcodes[['postalCode', 'geometry']], how='left', predicate='within'
)
df_crashes['ZIP CODE'] = df_crashes['ZIP CODE'].combine_first(
    crashes_with_zip.set_index(crashes_with_zip.index)['postalCode']
)

# -----------------------------
# 6. Final save
# -----------------------------
df_crashes.to_csv(FINAL_SAVE, index=False)
print("âœ… Final dataset saved:", FINAL_SAVE)
print("Remaining missing LAT/LON:", df_crashes[['LATITUDE','LONGITUDE']].isna().sum().to_dict())
print("Remaining missing BOROUGH:", df_crashes['BOROUGH'].isna().sum())
print("Remaining missing ZIP CODE:", df_crashes['ZIP CODE'].isna().sum())
```

```python
# Merge geocoded batches back, infer borough/zip from lat/lon, dedupe, save final
import pandas as pd
import os
import geopandas as gpd
from shapely.geometry import Point

SAVE_DIR = '/content/drive/MyDrive/geocoding_batches'
FINAL_SAVE = '/content/drive/MyDrive/df_crashes_geocoded_merged.csv'
FINAL_CLEAN = '/content/drive/MyDrive/df_crashes_geocoded_clean.csv'
FAILED_SAVE = '/content/drive/MyDrive/failed_geocodes.csv'

# 0. Quick checks
print("Original df_crashes shape:", df_crashes.shape)
print("Looking for batch files in:", SAVE_DIR)
batch_files = sorted([os.path.join(SAVE_DIR, f) for f in os.listdir(SAVE_DIR) if f.startswith('geocoded_batch_') or f.startswith('batch_')])
print(f"Found {len(batch_files)} batch files.")

if len(batch_files) == 0:
    raise SystemExit("No geocoded batch files found. Stop and check SAVE_DIR.")

# 1. Load & normalize batch files into a single DataFrame
batches = []
for f in batch_files:
    try:
        dfb = pd.read_csv(f)
    except Exception as e:
        print("WARNING: failed to read", f, ":", e)
        continue

    # Normalize possible column names
    # We want to end up with: ['index' (optional), 'COLLISION_ID' (optional), 'FULL ADDRESS' (optional), 'LATITUDE_GEOCODE', 'LONGITUDE_GEOCODE']
    colmap = {c: c.strip() for c in dfb.columns}
    dfb.rename(columns=colmap, inplace=True)

    # Some partial files use 'LATITUDE_GEOCODE' etc, others 'LATITUDE_GEOCODE.1' â€” find best matches
    lat_col = None
    lon_col = None
    for candidate in ['LATITUDE_GEOCODE','LATITUDE_GEOCODE.1','LATITUDE_GEOCODE ' ,'LAT','latitude','lat']:
        if candidate in dfb.columns:
            lat_col = candidate
            break
    for candidate in ['LONGITUDE_GEOCODE','LONGITUDE_GEOCODE.1','LONGITUDE_GEOCODE ','LON','longitude','lon']:
        if candidate in dfb.columns:
            lon_col = candidate
            break

    # If batch saved only 'LATITUDE'/'LONGITUDE' use them
    if lat_col is None:
        for candidate in ['LATITUDE','latitude','lat']:
            if candidate in dfb.columns:
                lat_col = candidate; break
    if lon_col is None:
        for candidate in ['LONGITUDE','longitude','lon']:
            if candidate in dfb.columns:
                lon_col = candidate; break

    # Identify id/index columns if present
    idx_col = None
    for candidate in ['index','Unnamed: 0','INDEX']:
        if candidate in dfb.columns:
            idx_col = candidate
            break

    id_col = None
    for candidate in ['COLLISION_ID','collision_id','collision id','collisionid']:
        if candidate in dfb.columns:
            id_col = candidate
            break

    addr_col = None
    for candidate in ['FULL ADDRESS','FULL_ADDRESS','FULLADDRESS','FULL ADDRESS ' ,'FULL_ADDRESS ' ,'FULLADDRESS ' ,'address','ADDRESS']:
        if candidate in dfb.columns:
            addr_col = candidate
            break

    # rename to standard columns
    rename_map = {}
    if lat_col: rename_map[lat_col] = 'LATITUDE_GEOCODE'
    if lon_col: rename_map[lon_col] = 'LONGITUDE_GEOCODE'
    if idx_col: rename_map[idx_col] = 'orig_index'
    if id_col: rename_map[id_col] = 'COLLISION_ID'
    if addr_col: rename_map[addr_col] = 'FULL ADDRESS'

    dfb = dfb.rename(columns=rename_map)
    # Keep only necessary columns
    keep_cols = [c for c in ['orig_index','COLLISION_ID','FULL ADDRESS','LATITUDE_GEOCODE','LONGITUDE_GEOCODE'] if c in dfb.columns]
    dfb = dfb[keep_cols].copy()

    # If orig_index exists but is read as float, make int where possible
    if 'orig_index' in dfb.columns:
        try:
            dfb['orig_index'] = pd.to_numeric(dfb['orig_index'], errors='coerce').astype('Int64')
        except:
            pass

    batches.append(dfb)

if not batches:
    raise SystemExit("No valid batches loaded after normalization.")

df_geocoded = pd.concat(batches, ignore_index=True, sort=False)
print("Concatenated geocoded rows:", len(df_geocoded))
print("Columns in geocoded df:", df_geocoded.columns.tolist())

# 2. Prepare a mapping to update df_crashes
# Priority updating strategy:
# 1) If batch stored original row index ('orig_index') â€” use that (most reliable)
# 2) Else if batch has 'COLLISION_ID' â€” use that
# 3) Else fallback to 'FULL ADDRESS' (take first lat/lon per address)

# Create Series placeholders for filling
lat_series = pd.Series(index=df_crashes.index, dtype='float64')
lon_series = pd.Series(index=df_crashes.index, dtype='float64')

# 2.a Use orig_index if present
if 'orig_index' in df_geocoded.columns:
    tmp = df_geocoded.dropna(subset=['orig_index'])
    # convert to int (drop NaN)
    tmp = tmp[tmp['orig_index'].notna()]
    tmp_idx = tmp['orig_index'].astype(int)
    # align length
    tmp = tmp.set_index(tmp_idx.index)  # keep index safe
    # create DataFrame mapped by orig_index -> lat/lon
    mapping = pd.DataFrame({
        'LATITUDE_GEOCODE': tmp['LATITUDE_GEOCODE'].values,
        'LONGITUDE_GEOCODE': tmp['LONGITUDE_GEOCODE'].values
    }, index=tmp_idx.values)
    # update series where index exists
    common_idx = mapping.index.intersection(df_crashes.index)
    lat_series.loc[common_idx] = mapping.loc[common_idx, 'LATITUDE_GEOCODE']
    lon_series.loc[common_idx] = mapping.loc[common_idx, 'LONGITUDE_GEOCODE']
    print("Mapped using orig_index:", len(common_idx))

# 2.b Use COLLISION_ID mapping next (only for rows still missing)
if 'COLLISION_ID' in df_geocoded.columns:
    # create first-lat-lon per COLLISION_ID (in case of duplicates)
    tmp2 = df_geocoded.dropna(subset=['COLLISION_ID'])
    tmp2 = tmp2[ tmp2['COLLISION_ID'].notna() ]
    tmp2['COLLISION_ID'] = tmp2['COLLISION_ID'].astype(str)
    first_by_id = tmp2.groupby('COLLISION_ID', as_index=False).first()[['COLLISION_ID','LATITUDE_GEOCODE','LONGITUDE_GEOCODE']]
    # make mapping dict
    lat_map = first_by_id.set_index('COLLISION_ID')['LATITUDE_GEOCODE'].to_dict()
    lon_map = first_by_id.set_index('COLLISION_ID')['LONGITUDE_GEOCODE'].to_dict()
    # apply where df_crashes still missing coords
    mask_missing = df_crashes['LATITUDE'].isna() | df_crashes['LONGITUDE'].isna()
    if 'COLLISION_ID' in df_crashes.columns:
        # convert to string to match mapping keys
        cid_series = df_crashes.loc[mask_missing,'COLLISION_ID'].astype(str)
        fill_lat = cid_series.map(lat_map)
        fill_lon = cid_series.map(lon_map)
        lat_series.loc[fill_lat[fill_lat.notna()].index] = fill_lat[fill_lat.notna()].values
        lon_series.loc[fill_lon[fill_lon.notna()].index] = fill_lon[fill_lon.notna()].values
        print("Mapped using COLLISION_ID:", fill_lat.notna().sum())

# 2.c Use FULL ADDRESS fallback (first lat/lon per address)
if 'FULL ADDRESS' in df_geocoded.columns:
    tmp3 = df_geocoded.dropna(subset=['FULL ADDRESS'])
    tmp3 = tmp3[tmp3['FULL ADDRESS'].notna()]
    # normalize address strings for matching
    tmp3['FULL ADDRESS NORM'] = tmp3['FULL ADDRESS'].str.strip().str.upper()
    first_by_addr = tmp3.groupby('FULL ADDRESS NORM', as_index=False).first()[['FULL ADDRESS NORM','LATITUDE_GEOCODE','LONGITUDE_GEOCODE']]
    lat_map_addr = first_by_addr.set_index('FULL ADDRESS NORM')['LATITUDE_GEOCODE'].to_dict()
    lon_map_addr = first_by_addr.set_index('FULL ADDRESS NORM')['LONGITUDE_GEOCODE'].to_dict()

    # select rows still missing in df_crashes
    mask_missing = df_crashes['LATITUDE'].isna() | df_crashes['LONGITUDE'].isna()
    addr_series = df_crashes.loc[mask_missing, 'FULL ADDRESS'].astype(str).str.strip().str.upper()
    fill_lat = addr_series.map(lat_map_addr)
    fill_lon = addr_series.map(lon_map_addr)
    lat_series.loc[fill_lat[fill_lat.notna()].index] = fill_lat[fill_lat.notna()].values
    lon_series.loc[fill_lon[fill_lon.notna()].index] = fill_lon[fill_lon.notna()].values
    print("Mapped using FULL ADDRESS fallback:", fill_lat.notna().sum())

# 3. Apply found lat/lon into df_crashes (only where df_crashes missing)
df_crashes['LATITUDE'] = df_crashes['LATITUDE'].combine_first(lat_series)
df_crashes['LONGITUDE'] = df_crashes['LONGITUDE'].combine_first(lon_series)

print("After merging batches, missing LAT:", df_crashes['LATITUDE'].isna().sum(), "missing LON:", df_crashes['LONGITUDE'].isna().sum())

# 4. Build GeoDataFrame from rows with lat/lon for spatial join
filled = df_crashes.dropna(subset=['LATITUDE','LONGITUDE']).copy()
print("Rows with coordinates for spatial join:", len(filled))

crashes_gdf = gpd.GeoDataFrame(
    filled,
    geometry=gpd.points_from_xy(filled['LONGITUDE'], filled['LATITUDE']),
    crs="EPSG:4326"
)

# 5. Spatial join to assign BOROUGH (use geometry column in boroughs DataFrame)
# Make sure boroughs and zipcodes GeoDataFrames are loaded and in EPSG:4326
if boroughs.crs != crashes_gdf.crs:
    boroughs = boroughs.to_crs(crashes_gdf.crs)
if zipcodes.crs != crashes_gdf.crs:
    zipcodes = zipcodes.to_crs(crashes_gdf.crs)

# perform sjoin; predicate='within' (geopandas >=0.10)
crashes_with_borough = gpd.sjoin(crashes_gdf, boroughs[['BoroName','geometry']], how='left', predicate='within')
# combine_back: set index = original index and then combine_first
crashes_with_borough_unique = crashes_with_borough[~crashes_with_borough.index.duplicated(keep='first')]
df_crashes['BOROUGH'] = df_crashes['BOROUGH'].combine_first(crashes_with_borough_unique['BoroName'])

# 6. Spatial join for ZIP code (postalCode column name may differ - adapt)
# Find postal code column name in zipcodes
zip_code_col = None
for c in zipcodes.columns:
    if 'postal' in c.lower() or 'zcta' in c.lower() or 'zip' in c.lower():
        zip_code_col = c
        break
if zip_code_col is None:
    raise SystemExit("Could not find ZIP/postal column in zipcodes GeoDataFrame. Check zipcodes.columns:", zipcodes.columns)

crashes_with_zip = gpd.sjoin(crashes_gdf, zipcodes[[zip_code_col,'geometry']], how='left', predicate='within')
crashes_with_zip_unique = crashes_with_zip[~crashes_with_zip.index.duplicated(keep='first')]
df_crashes['ZIP CODE'] = df_crashes['ZIP CODE'].combine_first(crashes_with_zip_unique[zip_code_col])

print("After spatial imputation - Missing BOROUGH:", df_crashes['BOROUGH'].isna().sum(), "Missing ZIP CODE:", df_crashes['ZIP CODE'].isna().sum())

# 7. Remove duplicates by COLLISION_ID if exists
if 'COLLISION_ID' in df_crashes.columns:
    before = len(df_crashes)
    df_crashes = df_crashes.drop_duplicates(subset='COLLISION_ID', keep='first')
    print(f"Removed {before - len(df_crashes):,} duplicate rows based on COLLISION_ID.")
else:
    print("No COLLISION_ID column found; skipping duplicates removal.")

# 8. Save intermediate artifacts
df_crashes.to_csv(FINAL_SAVE, index=False)
print("Saved merged dataset to:", FINAL_SAVE)

# Save cleaned (drop rows without coordinates) if you want a visualization-ready copy
df_final_clean = df_crashes.dropna(subset=['LATITUDE','LONGITUDE']).copy()
df_final_clean.to_csv(FINAL_CLEAN, index=False)
print("Saved cleaned (coords-only) dataset to:", FINAL_CLEAN)

# Save unresolved rows for manual review
failed = df_crashes[df_crashes['LATITUDE'].isna() | df_crashes['LONGITUDE'].isna()].copy()
failed.to_csv(FAILED_SAVE, index=False)
print("Saved failed/unresolved addresses to:", FAILED_SAVE)

# Summary
print("\nFINAL SUMMARY:")
print("Total rows in final df_crashes:", len(df_crashes))
print("Rows with coordinates (visualization-ready):", len(df_final_clean))
print("Unresolved rows saved:", len(failed))
```

```python
import geopandas as gpd
from shapely.geometry import Point
import pandas as pd

# Path to your cleaned CSV
CLEAN_PATH = '/content/drive/MyDrive/df_crashes_geocoded_clean.csv'

# Load the dataset
df_clean = pd.read_csv(CLEAN_PATH)
# Ensure the index is unique before proceeding
df_clean = df_clean.reset_index(drop=True)
print("Loaded df_clean shape:", df_clean.shape)

# 1ï¸âƒ£ Remove 0,0 coordinates
df_clean = df_clean[~((df_clean['LATITUDE'] == 0.0) & (df_clean['LONGITUDE'] == 0.0))].copy()
print("After removing 0,0 coords:", df_clean.shape)

# 2ï¸âƒ£ Convert to GeoDataFrame
gdf_clean = gpd.GeoDataFrame(
    df_clean,
    geometry=gpd.points_from_xy(df_clean['LONGITUDE'], df_clean['LATITUDE']),
    crs="EPSG:4326"
)

# Reproject to a projected CRS (meters) for accurate distance calculations
gdf_clean_proj = gdf_clean.to_crs(epsg=3857) # Use a projected version for spatial joins
boroughs_proj = boroughs.to_crs(epsg=3857)
zipcodes_proj = zipcodes.to_crs(epsg=3857)

# --- BOROUGH ---
# Perform sjoin for 'within' predicate
joined_boro_within = gpd.sjoin(gdf_clean_proj, boroughs_proj[['BoroName','geometry']], how='left', predicate='within')
# Drop duplicate indices from the sjoin result, keeping the first
joined_boro_within = joined_boro_within[~joined_boro_within.index.duplicated(keep='first')]

# Initialize a Series to store the final borough names, aligned with gdf_clean_proj's index
final_boroughs = pd.Series(index=gdf_clean_proj.index, dtype=object)
final_boroughs.update(joined_boro_within['BoroName']) # Update with 'within' results

# Identify remaining missing boroughs
missing_boro_idx = final_boroughs.isna()

if missing_boro_idx.any():
    print(f"Approximating {missing_boro_idx.sum()} BOROUGH values using nearest polygon...")
    gdf_to_approx = gdf_clean_proj.loc[missing_boro_idx] # Select only points still missing a borough
    if not gdf_to_approx.empty:
        nearest_boro = gpd.sjoin_nearest(
            gdf_to_approx,
            boroughs_proj[['BoroName','geometry']],
            how='left', max_distance=2000
        )
        # Drop duplicates for 'nearest' results as well
        nearest_boro = nearest_boro[~nearest_boro.index.duplicated(keep='first')]
        final_boroughs.update(nearest_boro['BoroName']) # Update final_boroughs

# --- ZIP CODE ---
joined_zip_within = gpd.sjoin(gdf_clean_proj, zipcodes_proj[['postalCode','geometry']], how='left', predicate='within')
joined_zip_within = joined_zip_within[~joined_zip_within.index.duplicated(keep='first')]

final_zipcodes = pd.Series(index=gdf_clean_proj.index, dtype=object)
final_zipcodes.update(joined_zip_within['postalCode'])

missing_zip_idx = final_zipcodes.isna()

if missing_zip_idx.any():
    print(f"Approximating {missing_zip_idx.sum()} ZIP CODE values using nearest polygon...")
    gdf_to_approx = gdf_clean_proj.loc[missing_zip_idx]
    if not gdf_to_approx.empty:
        nearest_zip = gpd.sjoin_nearest(
            gdf_to_approx,
            zipcodes_proj[['postalCode','geometry']],
            how='left', max_distance=2000
        )
        nearest_zip = nearest_zip[~nearest_zip.index.duplicated(keep='first')]
        final_zipcodes.update(nearest_zip['postalCode'])

# 5ï¸âƒ£ Assign back to df_clean
# Ensure assignment is done using the original index of df_clean
df_clean['BOROUGH'] = final_boroughs
df_clean['ZIP CODE'] = final_zipcodes

# 6ï¸âƒ£ Drop geometry column to save as regular DataFrame
# The original df_clean never had a 'geometry' column in this flow, so dropping it is not strictly necessary but harmless.
# If gdf_clean itself was modified directly, this line would be more critical.
if 'geometry' in df_clean.columns:
    df_clean = df_clean.drop(columns='geometry')

# 7ï¸âƒ£ Save cleaned dataset
FINAL_CLEAN_APPROX = '/content/drive/MyDrive/df_crashes_geocoded_clean_approx.csv'
df_clean.to_csv(FINAL_CLEAN_APPROX, index=False)
print("âœ… Cleaned dataset saved:", FINAL_CLEAN_APPROX)

# Quick stats
print("Missing BOROUGH:", df_clean['BOROUGH'].isna().sum())
print("Missing ZIP CODE:", df_clean['ZIP CODE'].isna().sum())
```
